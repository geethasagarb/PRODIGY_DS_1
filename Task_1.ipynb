Task 1: Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable , such as distribution of ages or genders in a population.

```python
#Importing necessary libraries

import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python

data['Fare'].hist(color='Green')

# Adding labels and title for better readability
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Histogram of Fare')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
data.boxplot(column='Fare', color='blue')

# Adding labels and title for better readability
plt.xlabel('Fare')
plt.ylabel('Value')
plt.title('Boxplot of Fare')

# Display the plot
plt.show()
```

```python
plt.scatter(data['Age'], data['Fare'], color='blue', edgecolors='black', linewidth=0.5)

# Adding labels and title for better readability
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

# Display the plot
plt.show()

```

```python

age_groups = data.groupby('Age')['Fare'].mean()  # You can use 'sum()' instead of 'mean()' if you prefer

# Creating a bar plot
age_groups.plot(kind='bar', color='blue', edgecolor='black')

# Adding labels and title for better readability
plt.xlabel('Age')
plt.ylabel('Mean Fare')
plt.title('Bar Plot of Mean Fare Across Age Groups')

# Display the plot
plt.show()
```
